```python
class Team():
    def __init__(self, team):
        self.name = team    
                
        teams = ['ATL', 'BOS', 'BRK', 'CHA', 'CHI', 'CLE', 'DAL', 'DEN', 'DET', 'GSW',
                 'HOU', 'IND', 'LAC', 'LAL', 'MEM', 'MIA', 'MIL', 'MIN', 'NOP', 'NYK',
                 'OKC', 'ORL', 'PHI', 'PHO', 'POR', 'SAC', 'SAS', 'TOR', 'UTA', 'WAS']
        
        east = ['ATL','BOS','BRK','CHA','CHI','CLE','DET','IND','MIA','MIL','NYK','ORL','PHI','TOR','WAS']
        west = ['DAL','DEN','GSW','HOU','LAC','LAL','MEM','MIN','NOP','OKC','PHO','POR','SAC','SAS','UTA']
        
        atlantic = ['TOR','BOS','PHI','BRK','NYK']
        central = ['MIL','IND','CHI','DET','CLE']
        southeast = ['MIA','ORL','CHA','WAS','ATL']
        northwest = ['MIN','OKC','DEN','POR','UTA']
        pacific = ['LAL','LAC','GSW','PHO','SAC']
        southwest = ['MEM','NOP','HOU','SAS','DAL']
        
        east_divisions = {'atlantic':atlantic,'central':central,'southeast':southeast}
        west_divisions = {'northwest':northwest,'pacific':pacific,'southwest':southwest}
        
        if self.name in east:
            self.conference = "east"
            self.conference_teams = east
            for div in east_divisions.keys():
                if self.name in east_divisions[div]:
                    self.division = div
                    self.division_teams = east_divisions[div]
        else:
            self.conference = "west"
            self.conference_teams = west
            for div in west_divisions.keys():
                if self.name in west_divisions[div]:
                    self.division = div
                    self.division_teams = west_divisions[div]
        
        self.opponents = []
        self.conference_opponents = []
        self.division_opponents = []
        self.conference_not_division = []
        
        for team in teams:
            if team != self.name:
                self.opponents.append(team)
                if team in self.conference_teams:
                    self.conference_opponents.append(team)
                    if team in self.division_teams:
                        self.division_opponents.append(team)
                    else:
                        self.conference_not_division.append(team)
```
